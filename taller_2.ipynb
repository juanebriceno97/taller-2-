```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

Primero es necesario importar los datos del precio de cierre de las acciones. Tambien es necesario incluir dentro de los tickers el del indice de S&P500 puesto que este nos servira para calcular los rendimientos y varianzas del mercado.

```python
tickers = ['^GSPC','AAPL','MSFT','AMZN','FB','TSLA','GOOGL','GOOG','BRK-B','JPM','JNJ','NVDA','V','UNH','HD','DIS','PG','MA','PYPL','BAC','INTC','ADBE','CMCSA','XOM','NFLX','VZ','CSCO','ABT','T','CRM','PFE']
data = pd.DataFrame()
for stock in tickers:
    data[stock] = wb.DataReader(stock,data_source='yahoo',start='2015-04-01',end='2021-04-16')['Close']
data.head(5)    
```

Ahora hallamos los pesos del portafolio.

```python
just_stocks = ['AAPL','MSFT','AMZN','FB','TSLA','GOOGL','GOOG','BRK-B','JPM','JNJ','NVDA','V','UNH','HD','DIS','PG','MA','PYPL','BAC','INTC','ADBE','CMCSA','XOM','NFLX','VZ','CSCO','ABT','T','CRM','PFE']
market_cap_data = wb.get_quote_yahoo(just_stocks)['marketCap']
Total_cap = sum(market_cap_data)
weights = market_cap_data / Total_cap
print(weights)
```

Ahora hallamos los retornos 

```python
returns = np.log(data/data.shift(1))
returns.head(5)
```

Hallamos la matriz de varianza covarianza y multiplicamos por 250 para tener varianzas y covarianzas anuales.

```python
cov = returns.cov()*250
cov
```

Si nos damos cuenta la covarianza de cada accion con el mercado esta en la primera columna de la matriz varianza-covarianza hallada. Podemos guardar estos valores en un array. 

```python
Cov_with_market = cov.iloc[1:31:,0]
Cov_with_market
```

Para calcular los betas solo nos falta la varianza del portafolio de mercado.

```python
market_var = returns['^GSPC'].var()*250
market_var
```

Calculando los Betas

```python
Betas = Cov_with_market/market_var
Betas
```

Calculamos el rendimiento y volatilidad del portafolio de mercado

```python
market_port_return = np.sum(returns['^GSPC'].mean()*weights)*250
market_port_volat = np.sqrt(market_var)
print(market_port_return)
print(market_port_volat)
```

Calculamos el rendimiento y volatilidad de nuestro portafolio

```python
portafolio_return = np.sum(returns.iloc[:,1:31].mean()*weights)*250
port_cov = returns.iloc[:,1:31].cov()*250
portafolio_volat = np.sqrt(np.dot(weights.T,np.dot(port_cov,weights)))
print(portafolio_return)
print(portafolio_volat)
```

calculamos el beta del portafolio

```python
port_Beta = np.sum(weights*Betas)
port_Beta
```

Ahora pasamos a la simulacion del portafolio

```python
## definimos el numero de portafolios
num_of_portafolios = 1000
##preparamos los arrays necesarios
number_of_symbols = len(tickers)-1

all_weights = np.zeros ((num_of_portafolios,number_of_symbols))

ret_arr = np.zeros(num_of_portafolios)

vol_arr = np.zeros(num_of_portafolios)

beta_arr = np.zeros(num_of_portafolios)

ratio_arr = np.zeros(num_of_portafolios)

#empezamos la simulacion

for i in range(num_of_portafolios):
    w = np.array(np.random.random(number_of_symbols))
    w = w / np.sum(w)
    
    all_weights[i, :] = w
    
    ret_arr[i] = np.sum((returns.iloc[:,1:31].mean()*w)*250)
    
    vol_arr[i] = np.sqrt(np.dot(w.T,np.dot(port_cov,w)))
    
    beta_arr[i] = np.sum(w*Betas)
    
    ratio_arr[i] = beta_arr[i] / ret_arr[i]
    
simulations_data = [ret_arr, vol_arr, beta_arr, ratio_arr, all_weights]

simulations_df = pd.DataFrame(data=simulations_data).T

simulations_df_columns = ['Returns', 'Volatility', 'Beta', 'B/R','Portafolio Weights']

simulations_df = simulations_df.infer_objects()

print('')
print('='*80)
print('SIMULATIONS RESULT')
print('-'*80)
print(simulations_df.head())
print('-'*80)
```

Finalmente encontramos dentro de nuestras simulaciones el portafolio con menor ratio beta/retorno

```python
min_ratio = simulations_df.loc[simulations_df[3].idxmin()]

print(min_ratio)
```
